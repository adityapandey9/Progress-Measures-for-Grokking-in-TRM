# Nanda Figure Walkthrough — TRM Minimal

Reproduce core grokking figures from shipped checkpoints (`nanda_faithful_50k`).

```python
# !pip install torch numpy matplotlib einops
import json
from pathlib import Path
import matplotlib.pyplot as plt

ROOT = Path(".")  # repo root after clone
CKPT = ROOT / "checkpoints/trm_minimal/seed_0/training_history.json"
```

```python
hist = json.loads(CKPT.read_text())
rows = [r for r in hist if str(r.get("step", "")).isdigit()]
steps = [int(r["step"]) for r in rows]
acc = [float(r["test_acc"]) for r in rows]
plt.figure(figsize=(8,3))
plt.plot(steps, acc)
plt.xlabel("Step"); plt.ylabel("Test accuracy"); plt.title("Grokking (seed 0)")
```
